## **2. Data transformation**

##### As already mentioned our target column will be a calculated column 'winner' which is an additional column we are creating.It is a categorical column containing two values '0' and '1'. The counties where Democrats are more popular  denoted as '0' and counties where Republicans are popular is denoted by '1'. Before that we have to compare our dataset with 'primary_results.csv' which we used for merging.

```python
#checking rows where both parties having zero votes and removing those rows.
data[((data["Republican"]==0)&(data["Democrat"]==0))]
```

```python
# removing rows with zero votes for both parties
data=data[~((data["Republican"]==0)&(data["Democrat"]==0))]
```

```python
data.shape
```

```python
#creating a column 'winner', where '0' and '1' represents winners democrat and Republican respectively
data["winner"]=np.where(data['Democrat'] > data['Republican'], 0, 1)
```

 - strong positive correlation between PST040210 and PST045214 and also between POP010210 WITH PST045214 and PST040210.
 - There is an inverse correlation of -0.59 between HSD310213(Persons per household 2009-13) and AGE775214 (percent of persons 65 years and above).
 - There is -0.68 correlation between HSG445213(home ownership rate) and HSG096213 (percentage of housing units in multi-unit structures).
 - We see a correlation of 0.89 between INC910213 (per capita monetary income in last 12 months) and INC110213(median household income).
 - a -0.73 correlation between INC910213 and PVY020213(percent of persons below poverty level) ; so dropping PVY020213 and INC110213.

## **b) Diamensionality Reduction**

###  **Principal component analysis**

```python
from sklearn.decomposition import PCA
pca = PCA().fit(X_std)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))

plt.xlabel('Number of Components', fontsize = 10)
plt.ylabel('Variance (%)', fontsize = 10) 
plt.title('The no.of Components needed to explain Variance', fontsize = 12)

plt.axhline(y=0.96, color='r', linestyle='dashed')
plt.show()
```

```python
pca = PCA(n_components = 25)
pca.fit(X_std)
x_pca = pca.transform(X_std)
x_pca.shape
```
